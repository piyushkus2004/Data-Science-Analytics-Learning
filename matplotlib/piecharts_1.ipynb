```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
labels = ["Sachin","Sehwag","Kohli","Yuvraj"]
runs = [18000,8000,12000,9500]

plt.title("Careers Runs of Indian Batsman")
plt.pie(runs,labels=labels)
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs,labels=labels,colors=colors)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'--'}
)
plt.show()
```

```python
explode = [0.1,0.5,0,0]

plt.pie(runs,labels=labels,explode=explode,colors=colors)
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode = [0.1,0.5,0,0]

plt.pie(runs,labels=labels,explode=explode,colors=colors,shadow=True)
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode = [0.1,0,0,0]

plt.pie(runs,labels=labels,explode=explode,colors=colors,shadow=True,startangle=90)
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode = [0.1,0,0,0]

plt.pie(runs,labels=labels,explode=explode,colors=colors,shadow=True,startangle=90,autopct='%1.1f%%',)
plt.title("Exploded Pie Example")
plt.show()
```
